# 정맥 인증 이미지 처리

### 필요한 라이브러리와 이미지 불러오기

```python
import cv2
import numpy as np
from PIL import Image
```

```python
img = cv2.imread("images\\20210511_184923_CROPPING.jpg")
```

### 이미지를 GRAY로 바꾸고 노이즈 제거 후 다시 BGR로 변환

```python
# noise
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
noise = cv2.fastNlMeansDenoising(gray)
noise = cv2.cvtColor(noise, cv2.COLOR_GRAY2BGR)
```

```python
image1 = Image.open("images\image1.jpg")
image1 = image1.resize((250, 250))
image1
```

### 히스토그램 균등화

```python
# equalist hist
kernel = np.ones((7,7),np.uint8)
img = cv2.morphologyEx(noise, cv2.MORPH_OPEN, kernel)
img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
img_output = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)
```

```python
image2 = Image.open("images\image2.jpg")
image2 = image2.resize((250, 250))
image2
```

### 이미지 반전 처리

```python
# invert
inv = cv2.bitwise_not(img_output)
```

```python
image2 = Image.open("images\image3.jpg")
image2 = image2.resize((250, 250))
image2
```

### 이미지 침식 처리

```python
# erode
gray = cv2.cvtColor(inv, cv2.COLOR_BGR2GRAY)
erosion = cv2.erode(gray,kernel,iterations = 1)
```

```python
image2 = Image.open("images\image4.jpg")
image2 = image2.resize((250, 250))
image2
```

### 이미지 골격화

```python
# skel
img = gray.copy()
skel = img.copy()
skel[:,:] = 0
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (5,5))
iterations = 0

while True:
    eroded = cv2.morphologyEx(img, cv2.MORPH_ERODE, kernel)
    temp = cv2.morphologyEx(eroded, cv2.MORPH_DILATE, kernel)
    temp  = cv2.subtract(img, temp)
    skel = cv2.bitwise_or(skel, temp)
    img[:,:] = eroded[:,:]
    if cv2.countNonZero(img) == 0:
        break
```

```python
image2 = Image.open("images\image5.jpg")
image2 = image2.resize((250, 250))
image2
```

### 정맥을 더 잘 보이게 하기 위한 임계 값 적용

```python
ret, thr = cv2.threshold(skel, 5,255, cv2.THRESH_BINARY);
```
